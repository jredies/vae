```python
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt

import glob
```

```python
!ls logs/aug/stats*.csv
```

```python
stats = {}

for x in glob.glob('logs/aug/stats_*.csv'):
    # print(x)
    
    layers = x.split('-')[1][0]
    geo = x.split('_geo-')[1].split('_')[0]
    latent_factor = x.split('latent_f-')[1].split('.csv')[0]

    train_loss = pd.read_csv(x).iloc[-1].train_loss
    val_loss = pd.read_csv(x).iloc[-1].val_loss

    if layers != '5':
        stats[(int(layers), geo, latent_factor, 'train_loss')] = train_loss
        stats[(int(layers), geo, latent_factor, 'val_loss')] = val_loss

```

```python
srs = pd.Series(stats).sort_index()

srs.index.names = ['layers', 'geometry', 'latent_factor', 'measure']

srs = srs.unstack('latent_factor')
```

```python
val = srs.xs(key='val_loss', level='measure')
train = srs.xs(key='train_loss', level='measure')
```

```python
(train/val).to_excel('perce.xlsx')
(train/val)
```

```python
val.to_excel('val.xlsx')
val
```

```python

sns.set_theme()


fig, ax = plt.subplots(1, 3, figsize=(30, 7), sharey=True)

col = ['train_loss', 'val_loss']

base = pd.read_csv("logs/aug/exp_1.0.csv").iloc[5:].loc[:, col]

base.plot(ax=ax[0], title='No Learning Rate Schedule')

plateau = pd.read_csv("logs/aug/plateau_0.25.csv").iloc[5:].loc[:, col]
plateau.plot(ax=ax[1], title='Plateau Learning Rate Schedule Gamma=0.25 Plateau Patience 5')

exp = pd.read_csv("logs/aug/exp_0.97.csv").iloc[5:].loc[:, col]
exp.plot(ax=ax[2], title='Exponential Decay Gamma=0.97')

ax[0].axhline(95, color='r', linestyle='--', label='95 loss level')
ax[1].axhline(95, color='r', linestyle='--', label='95 loss level')
ax[2].axhline(95, color='r', linestyle='--', label='95 loss level')


ax[0].legend()
ax[1].legend()
ax[2].legend()


plt.savefig('learning_rate_schedule.pdf')
```
